```python
#in this analysis I decided to look the relationship between SWD status and graduation rate"
import pandas as pd
import numpy as np
import seaborn as sns
grad = pd.read_csv('/content/2017_-_2018_Graduation_Outcomes_School_20240102.csv')
pd.options.display.max_rows = 100
grad.head(100)
```

```python
# Checking data types
grad.dtypes
```

```python

grad['Cohort'] = grad['Cohort'].astype(str)
grad['Demographic Variable'] = grad['Demographic Variable'].astype(str)
```

```python
grad.isnull().values.any()
```

```python
grad.loc[grad['Total Grads % of cohort']== 's']
```

```python
# removed rows with 's' which represented small student number
grad = grad.drop(grad[grad['Total Grads #'] == 's'].index)
grad = grad.drop(grad[grad['Total Grads % of cohort'] == 's'].index)
grad = grad.drop(grad[grad['Total Cohort #'] == 's'].index)
```

```python
# changing data types
grad['Total Grads % of cohort']= grad['Total Grads % of cohort'].apply(pd.to_numeric)
```

```python
grad['Total Cohort #'] = grad['Total Cohort #'].astype(str)
```

```python
dft = (grad[grad['Total Cohort #'].str.contains(',')]).copy()
dft
```

```python
grad['Total Cohort #']= grad['Total Cohort #'].str.replace(',','')
```

```python
grad['Total Cohort #']= grad['Total Cohort #'].apply(pd.to_numeric)
```

```python
m = grad.rename(columns={'Total Grads #':'TGN' }).copy()
```

```python
m.dtypes
```

```python
m['TGN'] = m['TGN'].astype(str)
```

```python
m['TGN']= m['TGN'].str.replace(',','')
```

```python
#reduced rows to only SWD vs Not SWD
t = (m[m['Demographic Variable'].str.contains('SWD')])
```

```python
t['Cohort'].unique()
```

```python
#dropped to have just two categories 4 year and 6 year graduatipon rates
t = t.drop(t[t.Cohort.isin(['5 year June','4 year August','5 year August'])].index)
```

```python
# just used single school to see what data was appearing for each school
dfO = (t[t['School Name'].str.contains('ORCHARD COLLEGIATE ACADEMY')])
df2011Orchard = (dfO[dfO['Cohort Year']== 2011])

df2011Orchard
```

```python
# some decriptives looking ta mean size of cohorts and graduation rates for all years s
t.describe()
```

```python
# looking at graduation rates rates and total size of cohort for all years both 4 year and 6 year graduation rate
dfg = t.groupby(['Cohort','Cohort Year','Demographic Variable'], as_index=False).agg({'Total Grads % of cohort':'mean','TGN':'sum','Total Cohort #':'sum','School Total':sum, 'Class Size':'mean'}).copy()
dfg
```

```python
dfg['gradper'] = (dfg['TGN']/dfg['Total Cohort #']).copy()
dfg
```

```python
#plot to see changes over time in graduation rates for student with and without diabilies
sns.lineplot(data=dfg, x="Cohort Year", y="gradper", hue="Demographic Variable", style="Cohort")
```

```python
dfresults = dfg.groupby(['Cohort','Demographic Variable']).agg({'gradper':'mean', 'Total Cohort #':'sum','Class Size':'mean'})
print(dfresults)
```

```python
dfp =dfg.copy()
dfp = dfp.drop(dfp[dfp.Cohort.isin(['4 year June'])].index)
dfp['Cohort Year'] = dfp['Cohort Year'].astype(str)
values=["2012","2001"]
df2012_2021 = (dfp[dfp["Cohort Year"].isin(values)] ).copy()
df2012_2021['Percent Graduduated in 6 year'] = (df2012_2021['TGN']/df2012_2021['Total Cohort #']*100)
df2012_2021.head(25)
```

```python
df2012_2021.dtypes
```

```python
dfcompg = df2012_2021.groupby(['Cohort Year','Demographic Variable'], as_index=  False).agg({'Total Cohort #':'sum','Total Grads % of cohort':'mean','TGN':'sum', 'Percent Graduduated in 6 year':'mean'})
dfcompg
```

```python
p = sns.barplot(data=dfcompg, x="Demographic Variable", y="Total Cohort #", hue="Cohort Year")
for h in p.containers:
     p.bar_label(h)
```

```python
CP = sns.catplot(
    dfcompg, kind="bar",
    x="Cohort Year", y="Total Cohort #", col="Demographic Variable", aspect= .7)
for CP in CP.axes.ravel():
  for c in CP.containers:
    CP.bar_label(c)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
centered=dfg.copy()
centered['Cohort Year']= centered['Cohort Year'].apply(pd.to_numeric)
centered.dtypes
```

```python

center_function = lambda x: x - x.mean()
data = pd.DataFrame({ 'x': centered['Demographic Variable'], 'y': centered['Total Grads % of cohort'], 'z':centered['Cohort Year'], 'f':centered['Total Cohort #']}).copy()
centered['Cohort Year'] = center_function(centered['Cohort Year'])
model = ols("y ~ x*z", data).fit()

print(model.summary())
```
